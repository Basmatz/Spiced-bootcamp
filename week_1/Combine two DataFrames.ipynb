```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
lifeexp = pd.read_excel('gapminder_lifeexpectancy.xlsx')
lifeexp
```

```python
year=lifeexp[2010]
year
```

```python
lifeexp2=lifeexp.rename(columns={"Life expectancy": "country"})
lifeexp2
```

```python
cont= pd.read_csv('continents.csv', sep = ";")
cont
```

```python
cont_year=year.append([cont])
cont_year
```

```python
df=pd.merge(lifeexp2, cont, on=['country'], how='inner')
df
```

```python
df=df.dropna()
df
```

```python
df.groupby(['continent']).min()   # print the highest and lowest life expectancy for each continent
```

```python
df.groupby(['continent']).max()
```

```python
df.groupby(['continent']).max().plot(figsize=(12,10))  # plot the numbers
```
