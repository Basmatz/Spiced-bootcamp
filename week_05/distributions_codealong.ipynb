```python
import numpy as np
from PIL import Image #Python image library
from matplotlib import pyplot as plt
```

```python
a = np.zeros((400,150), dtype=np.uint8) #uint8 = integers, u=unsigned, 8bits
xcoord = np.random.randint(0,399, size=300)
ycoord = np.random.randint(0,149, size=300)

a[xcoord, ycoord] = np.random.randint(0,255, size=300) # random brightness

Image.fromarray(a.T).resize((800, 300))
```

```python
b = np.random.randint(0,255, size=(400, 150), dtype=np.uint8)
Image.fromarray(b.T).resize((800,300))
```

We want to know the distribution of the dot colours:


```python
plt.hist(b.flatten(), bins=20)
```

This is a *uniform distribution* - each number has a an equal probability of appearing


```python
c = np.random.random(size = 100000)
plt.hist(c, bins=20)
```

## Normal (Gaussian) distribution

```python
d = np.random.normal(loc=128, scale=30, size=(100000))
p=plt.hist(d, bins=50)
```

```python
d = np.random.normal(loc=128, scale=30, size=[400,150])

d = d.round().astype(np.uint8)
Image.fromarray(d.T).resize((800, 300))
```

There are fewer white/black points because the distribution has most points in the middle

```python
a = np.zeros((400,150), dtype=np.uint8) #uint8 = integers, u=unsigned, 8bits

xcoord = np.random.normal(loc=200, scale=30, size=300)
xcoord = xcoord.round().astype(np.uint8)

ycoord = np.random.randint(0,149, size=300)

a[xcoord, ycoord]= np.random.randint(0,255, size=300) # random brightness

Image.fromarray(a.T).resize((800, 300))
```

## Triangular Distribution
 - equivalent of rolling two dice and adding their values

```python
t= np.random.triangular(2,7,12, size=100000).round()
pp= plt.hist(t, bins=11)
```

```python
r = np.random.randint(0,6, size=(100, 2))
rsum = r.sum(axis=1)
```

```python
plt.hist(r, bins=11)
plt.hist(rsum, bins=11)
```

```python
rmore = np.random.randint(0,6, size=(1000000, 100))
ppp= plt.hist(rmore.sum(axis=1), bins=50)
```

## Central Limit Theorem

The more uniform distributions you add together, the more the sum approaches the normal distribution - even if the distibutions are not the same 


## Binomial Distribution

```python
# Toss a coin 10 times and get heads three times
coins = np.random.binomial(n=100, p=0.5, size=100000)
```

```python
coins[:10]
```

```python
import pandas as pd
pd.Series(coins).value_counts().sort_index().plot.bar()
```

At higher n this also approaches the gaussian distribution (but discretised), with $p\neq0.5$ the distribution is skewed

## Poisson distribution

*in dancing schools, a single falls in love every 11 minutes*
- a rare event with an average frequency
- we assume that the events are independent

```python
singles_per_minute = 1/11
singles_per_hour = 60/11
p = np.random.poisson(lam=singles_per_hour, size=10000)
```

```python
ppp = plt.hist(p, bins=18)
```
