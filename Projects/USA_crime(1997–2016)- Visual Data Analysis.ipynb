```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

### Crime in the United States, by Volume and Rate per 100,000 Inhabitants, 1997â€“2016 (FBI: UCR)

```python
df=pd.read_excel('FBI-crime-in-the-u.s.xls', skiprows=3)
df=df.head(20)
df=df.replace({'20015':2001, '20156':2015})
df
```

```python
crime=df[['Year','Population1','Violent\ncrime2','Burglary']]
crime
```

```python
x= crime['Year']
y= crime['Burglary']
y2=crime['Violent\ncrime2']
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y , color= 'green')
plt.plot(x,y2 , color= 'red')
plt.xlabel('year')
plt.ylabel('amount')
plt.title('Crime in the United States, FBI:UCR')
plt.legend(['Burglary', 'Violent\ncrime2'])
```

### the violent crimes a bit more deeply

```python
df=df[['Year','Murder and\nnonnegligent \nmanslaughter','Rape\n(legacy \ndefinition4)']]
df
```

```python
x= df['Year']
y= df['Murder and\nnonnegligent \nmanslaughter']
y2= df['Rape\n(legacy \ndefinition4)']
plt.figure(figsize=(7,4))
plt.scatter(x,y , color= 'green')
plt.scatter(x,y2 , color= 'red')
plt.xlabel('year')
plt.ylabel('amount')
plt.title('Crime in the United States, FBI:UCR')
plt.legend(['Murder and\nnonnegligent \nmanslaughter', 'Rape\n(legacy \ndefinition4'])
```

 ** Rape\n(legacy \ndefinition4) ** The figures shown in this column for the offense of rape were estimated using the legacy UCR definition of rape. See data declaration for further explanation.

** Murder and\nnonnegligent \nmanslaughter ** The murder and nonnegligent homicides that occurred as a result of the events of September 11, 2001, are not included in this table. 

```python
rape=df['Rape\n(legacy \ndefinition4)'].mean()
rape
```
