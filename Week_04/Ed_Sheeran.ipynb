```python
songs = soup.find_all('td', {'class':'tal qx'})
```

```python
links = list()
for td in soup.find_all('td'):
    if "tal" in td.get('class',[]):
          links.append('https://www.lyrics.com'+td.find('a')['href'])
```

```python
for i in range(len(links)): 
    url = links[i]
    temp_req = requests.get(url)
    pattern = "\/lyric\/.+\/.+\/(.+)$"
    with open("EdSheeran/EdSheeran--" + re.findall(pattern, links[i])[0] + '.txt','w') as file:
        soup_Shakira = BeautifulSoup(temp_req.text)
        lyrics = soup.pre.get_text()
        file.writelines(lyrics)
file.close()
```

```python
corpus = list()
for i in range(100):
    with open("EdSheeran/EdSheeran--" +re.findall(pattern, links[i])[0] + '.txt','r') as reader:
        doc= reader.read()
        doc.lower()
        doc.split()
        reader.close
        corpus.append(doc)
```

```python
Artist_name = list()
for i in range(100):
    name = print('Ed Sheeran')
    Artist_name.append(name)
```

```python
corpus
```

```python
cv = CountVectorizer()

```

```python
cv.fit(corpus)
```

```python
vec_corpus = cv.transform(corpus)
```

```python
vec_corpus.todense()
```

```python

df = pd.DataFrame(vec_corpus.todense(), columns=cv.get_feature_names())
df
```
