```python
for year in list(range(1960,2016)) :
    
 df_year = df3.unstack(1)
 df_year = df_year[year]
 df_year = df_year.unstack(1)

 x = df_year['fertility']
 y = df_year['lifeexp']
 sizes = df_year['population'] * ( 10 ** -5 )

 fig = plt.scatter(x, y,
            s = sizes,
            c = cmap,    
            alpha=0.7)

 plt.axis([1, 9, 20, 80])
 plt.xlabel('Fertility')
 plt.ylabel('Life Expectancy')
 plt.savefig('lifeexp_' + str(year) + '.png', dpi = 150)
 plt.clf() 
    
    
    # execute more than one time. there is just one picture for one year. why? 
    # because of the same name !!!!
```
