```python
import pandas as pd
```

```python
life_expectancy_data = pd.read_excel("gapminder_lifeexpectancy.xlsx", index_col = 0)

life_expectancy = pd.DataFrame(life_expectancy_data)

```

```python
life_expectancy = life_expectancy[[1950, 1975, 2000, 2015]]
```

```python
import matplotlib.pyplot as plt

life_expectancy[2015].hist()
plt.savefig('histo.png')
```

```python
plt.title('Histogram of Life expectancies')
plt.xlabel('Life extectancy in years')
plt.ylabel('Frequency')
plt.axis([25.0, 85.0, 0.0, 30.0])
life_expectancy[1950].hist(bins=20, facecolor = '#ff0000', alpha=0.3)
life_expectancy[2000].hist(bins=20)
life_expectancy[2015].hist(bins=20)

plt.savefig('histo.png', dpi=150)
```
