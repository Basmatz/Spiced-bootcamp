```python
import numpy as np
from PIL import Image # Python Imaging Library
from matplotlib import pyplot as plt
```

```python
a = np.zeros((400, 150), dtype=np.uint8)
```

creates a numpy array of zeroes, uint8 means unsigned (cannot be negative), integer number with eight binary digits (bits)

```python
a.shape
```

```python
# create a 400 x 150 array with random values 0..255
b = np.random.randint(0, 255, size=(400, 150), dtype=np.uint8)
Image.fromarray(b.T).resize((800, 300))
```

```python
b.mean()
```

```python
plt.hist(b.flatten(), bins=20)
```

```python
xcoord = np.random.randint(0, 399, size=(300)) # create 300 random positions for x in range 0-399
```

```python
ycoord = np.random.randint(0, 149, size=(300)) # create 300 random positions for y in range 0-149
```

```python
a[xcoord, ycoord] = 255 # brightness
```

```python
Image.fromarray(a.T).resize((800, 300)) # resolution // .T is a transpose
```

### Uniform distribution

```python
c = np.random.random(size=(10000))
plt.hist(c, bins=20)
```

### Normal or Gaussian Distribution

```python
d = np.random.normal(loc=128.0, scale=30.0, size=(100000)) # loc is the mean
```

```python
p = plt.hist(d, bins=50)
```

```python
min(d), max(d)
```

```python
d = np.random.normal(loc=128.0, scale=30.0, size=(400,150))
d = d.round().astype(np.uint8)
Image.fromarray(d.T).resize((800, 300))
```

### Triangular Distribution

```python
t = np.random.triangular(2, 7, 12,size=(10000)).round()
```

```python
plt.hist(t)
```
