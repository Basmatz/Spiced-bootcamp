# STEP 10: Scatter plots 1960-2015 / STEP 11: Adjusting population size

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pwd
```

```python
year_list = list(range(1960,2016))
df_in = df3.unstack(1)

def makeplot (df3, year):
    
    def_temp = df_in[year]
    def_temp = def_temp.unstack(1)
    
    x = def_temp['fertility']
    y = def_temp['lifeexp']
    sizes = def_temp['population']/10**5 # in 100k units (otherwise will overwhelm the scale)
    cmap = plt.get_cmap('Pastel2', lut = len(def_temp)).colors

    
    fig = plt.scatter(x, y,
                s = sizes,
                c = cmap,   
                alpha=0.3)
    plt.axis([1, 10, 20, 80])
    plt.xlabel('Fertility')
    plt.ylabel('Life Expectancy (Years)')
    plt.title(str(year))
    
    plt.savefig('scatter_' + str(year) + '.png', dpi = 150)
    plt.clf() 
    
    return

for date in year_list:
    makeplot(df3, date)
```
