# Fizz Buzz Challenge

My solution:

```python
for fizzbuzz in range(1,101):
    if fizzbuzz % 3 == 0 and fizzbuzz % 5 == 0:
        print(str(fizzbuzz) + " fizzbuzz")
    elif fizzbuzz % 3 == 0:
        print(str(fizzbuzz) + " fizz")
    elif fizzbuzz % 5 == 0:
        print(str(fizzbuzz) + " buzz")

```

Someone else's solution:

```python
def func(num):
    for i in range(1,num+1):
        print(i, end=" ")
        # this prints each number and a newline, with outcome of fizzbuzz next to it if applicable
        if (i%3 == 0) & (i%5 == 0):
            print("FizzBuzz")
        elif i%3 == 0:
            print("Fizz")
        elif i%5 == 0:
            print("Buzz")
        else:
            print()
func(100)
```
