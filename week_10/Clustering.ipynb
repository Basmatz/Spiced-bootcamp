# Clustering

```python
from sklearn.datasets import make_moons, make_blobs
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# X, y = make_moons(50, noise=0.1, random_state=4)
X, y = make_blobs(50, n_features=2, random_state=42)

colors = np.array(['blue', 'red', 'green'])

plt.scatter(X[:,0], X[:,1], c=colors[y])
```

### K-Means

```python
from sklearn.cluster import KMeans
```

```python
m = KMeans(n_clusters=3)
m.fit(X)
ypred = m.predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=colors[ypred])
```

### Agglomerative Clustering

```python
import seaborn as sns
```

```python
sns.clustermap(X.T)
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN

X, _ = make_moons(1000, noise=0.05)
#outliers = np.random.normal(0.0, 1.0, (10, 2))

#x = np.concatenate([X, outliers])


m = DBSCAN(eps=0.2)
m.fit(X)

c = m.fit_predict(X)


c = m.labels_.astype(np.int)

plt.scatter(X[:,0], X[:,1], c=colors[c])
```

```python
mod = KMeans(n_clusters=2)
mod.fit(X)

c = mod.fit_predict(X)


c = mod.labels_.astype(np.int)

plt.scatter(X[:,0], X[:,1], c=colors[c])
```
