```python
import numpy as np
from PIL import Image  #Python Imaging Library
from matplotlib import pyplot as plt
```

```python
a = np.zeros((400, 150), dtype=np.uint8)     # unsigned integers numbers 
xcoord = np.random.randint(0, 399, size=(300))
ycoord = np.random.randint(0, 149, size=(300))

a[xcoord, ycoord] = np.random.randint(0, 255, size=(300))

Image.fromarray(a.T).resize((800, 300))
```

```python
# to create a 400 x 150 array with random values 0..244
b = np.random.randint(0, 215, size=(400, 150), dtype =np.uint8)
Image.fromarray(b.T).resize((800, 300))
```
