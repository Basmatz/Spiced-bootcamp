# Gapminder Histogram

## Step 0: Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#jupyter specific: this will show plots in line for jupyter notebooks
```

## Step 1: Read in data

```python
life = pd.read_excel('gapminder_lifeexpectancy.xlsx', index_col=0)
# life.head(5)
# life.info()
```

## Step 2: Select several columns by year

```python
selected = life[[1931, 1966, 1988]] #birth years: my grandmother, mother, me
```

Note to self -- because col names are INTEGERS, must select using `[[1988]]` syntax i.e. double sq brackets.

## Step 3: Draw Histogram

```python
selected[[1931]].hist()
plt.savefig('hist_1928.png')
```

## Step 4: Play with Bin Size

Stacking plots: use `selected[1928]` syntax (1 set of sq brackets)

```python
selected[1931].hist(bins=20)
selected[1966].hist(bins=20)
selected[1988].hist(bins=20)
```

Individual plots: use `selected[[1928]]` syntax (2 sets of sq brackets)

```python
selected[[1931]].hist()
selected[[1966]].hist()
selected[[1988]].hist()
```

## Steps 5-7: Make it pretty :)

```python
bins = 20
alpha = .6
palette = ['#96ceb4', '#ffeead', '#ffcc5c'] # "beach colours palette"
font = {'fontname': 'Helvetica'}            # source: https://stackoverflow.com/questions/21321670/how-to-change-fonts-in-matplotlib-python


selected[1931].hist(bins=bins,
                    facecolor=palette[0],
                    alpha=alpha,
                    histtype='bar',)   #Why do we have to repeat these here? any way to generate in one step?
selected[1966].hist(bins=bins,
                    facecolor=palette[1],
                    alpha=alpha,
                    histtype='bar',)
selected[1988].hist(bins=bins,
                    alpha=alpha,
                    histtype='bar',)


plt.title('Life expectancy across the 20th century', **font)
plt.xlabel('Age in years', **font)
plt.ylabel('No. countries', **font)

## Set the font name for axis tick labels -- source: http://jonathansoma.com/lede/data-studio/matplotlib/changing-fonts-in-matplotlib/
#for tick in ax.get_xticklabels():
#    tick.set_fontname(font)
#for tick in ax.get_yticklabels():
#    tick.set_fontname(font)


plt.axis([25.0, 80.0, 5.0, 55.0])
plt.savefig('final_hist.png', dpi=300)
```
