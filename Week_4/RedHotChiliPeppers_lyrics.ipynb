```python
links = list() 
for td in soup.find_all('td'):
    if "tal" in td.get('class',[]):
          links.append('https://www.lyrics.com'+td.find('a')['href'])
```

```python
for i in range(len(links)): 
    url = links[i]
    temp_req = requests.get(url)
    songpattern = "\/lyric\/.+\/.+\/(.+)$"
    with open("RedHotChiliPeppers" + re.findall(songpattern, links[i])[0]+ '.txt','w') as file:
        soup = BeautifulSoup(temp_req.text)
        lyrics = soup.pre.get_text()
        file.writelines(lyrics)
file.close()


```
