```python
# Make a for loop to get a list with  links of the songs
links = list() 
for td in BSoup.find_all('td'):
    if "tal" in td.get('class',[]):
          links.append('https://www.lyrics.com'+td.find('a')['href'])
         
```

```python
# Creating a for loop to extract each of the lyrics and save it into a file

for i in range(len(links)): 
    url = links[i]
    temp_req = requests.get(url)
    pattern = "\/lyric\/.+\/.+\/(.+)$"
    with open("shakira/Shakira--" + re.findall(pattern, links[i])[0] + '.txt','w') as file:
        soup_Shakira = BeautifulSoup(temp_req.text)
        lyrics = soup_Shakira.pre.get_text()
        file.writelines(lyrics)
file.close()
```
