lets start with feature engineering

drop casual and registered as they are included in count

what is atemp? "feels like" temperature -a quantification of your reaction and the worlds reaction to the temperature situation during a given condition. There a lot of factors to consider, a list of them would be like this: Temperature,Pressure,Humidity,Wind,Sunlight,ETC.

Will drop atemp as there are humidity,wind and temp columns

It seems like the season is not important for the users, the counts are very similar in each weather category

```python
def create_time_col(df):
    
    df=df.copy()
   
    df['year']= pd.to_datetime(df['datetime']).dt.year
    df['month']= pd.to_datetime(df['datetime']).dt.month
    df['day']= pd.to_datetime(df['datetime']).dt.day
    df['hour']= pd.to_datetime(df['datetime']).dt.hour
    return df
```

```python
one_hot_features =['hour','weather','workingday','holiday']
date_feature =['datetime']
polynomial_features =['workingday','windspeed','humidity','hour','temp']

passing_features = ['season','year','day','temp','windspeed','humidity']
```

```python
column_transformer = ColumnTransformer(transformers=[
                        ('onehot', OneHotEncoder(), one_hot_features),
                        ('interaction_polynomial', PolynomialFeatures(interaction_only=True), polynomial_features),
                        ('pass', 'passthrough', passing_features)
])
```

```python
transformer_pipeline = Pipeline(steps=[('create_time_col', FunctionTransformer(create_time_col)),
                        ('column_transformer', column_transformer)])
```

```python
X_train_transformed = transformer_pipeline.fit_transform(X_train)
```

```python
# Linear Regression
```

```python
m=LinearRegression()
```

```python
m.fit(X_train_transformed, y_train)
```

```python
m.score(X_train_transformed, y_train)
```

```python
# Cross validation
```

```python
cross_val_score(m, X_train_transformed, y_train)
```
